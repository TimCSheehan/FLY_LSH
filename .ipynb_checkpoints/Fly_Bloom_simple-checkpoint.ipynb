```python
# create input data to be hashed
# should be uniform of size arbitrily larger than 350
dIn = 500
nIn = 5000
m = 350 # nKC to a'3
k = 17
datIn = np.random.uniform(-1,1,(nIn,dIn)) # was [-1, 1]
```

```python
# create projection matrix (dense Gaussian)
M = np.random.randn(dIn,m)

# create projection matrix (sparse binary)
num_sample = 6
MSB6 = np.zeros((dIn,m))
for row in range(dIn):
    idx = random.sample(range(m), num_sample) 
    MSB6[row,idx] = 1

datHash = np.dot(datIn,M)
datHashSB = np.dot(datIn,MSB6)
# sparsify output    
def sparsify_rows(space,n_keep):
    nrow = np.shape(space)[0]
    outHash = np.zeros(np.shape(space))
    inds = np.arange(nrow)
    for i in range(nrow):
        this_line = space[i]
        val_tuple = list(zip(this_line,inds))
        this_hash = sorted(val_tuple)[-n_keep:]
        this_hash = [ind[1] for ind in this_hash]
        outHash[i,this_hash] = 1
    return outHash

outHash = sparsify_rows(datHash,k)
outHashSB = sparsify_rows(datHashSB,k)
```

```python
def bloomFP(n=20,n_chk = 1):
    val_use = np.random.choice(nIn,size=n+n_chk,replace=False)
    bloom = np.any(outHash[val_use[:n]],axis=0) # to go along columns

    checker = outHash[val_use[-n_chk:]] # n_chk x m
    mtchs = [np.all(bloom[check==1]) for check in checker]
    return np.mean(mtchs)

def bloomFP_tol(n=20,n_chk = 1,tol=10):
    val_use = np.random.choice(nIn,size=n+n_chk,replace=False)
    bloom = np.any(outHash[val_use[:n]],axis=0) # to go along columns

    checker = outHash[val_use[-n_chk:]] # n_chk x m
    mtchs = [np.sum(bloom[check==1])>=tol for check in checker]
    return np.mean(mtchs)
```

```python
# compute probabilities of true match
from functools import reduce
import operator as op
def ncr(n, r):
    r = min(r, n-r)
    if r == 0: return 1
    numer = reduce(op.mul, np.arange(n, n-r, -1))
    denom = reduce(op.mul, np.arange(1, r+1))
    return numer//denom
#def factorial(n):return reduce(lambda x,y:x*y,[1]+range(1,n+1))
def pTP(m,k,tol):
    if tol == 0:
        mf = 1 # mult factor
    else:
        mf = reduce(op.mul,np.arange(k, k-tol, -1))
    prob = mf/ncr(m,k)
    return prob
def hamming_distance(x1,x2):
    assert(len(x1)==len(x2))
    try:
        return sum(x1^x2) # ^ is xor
    except:
        return sum((x1>0)^(x2>0)) # ^ is xor
        
```

```python
# make clones of a value, perterb
clone = outHash[-1]
hamming_distance(clone,outHash[0]) # hamming distance = 0
# now lets perterb
def perterb_hash(brown,n_perterb):
    hashbrown = brown.copy()
    keys = np.where(hashbrown==1)[0]
    assert(len(keys)>=n_perterb)
    key_perterb = np.random.choice(keys,n_perterb,replace=False)
    noMove = keys.copy()
    for i in range(n_perterb):
        randInt = np.random.randint(0,len(hashbrown))
        while any(noMove == randInt):
            randInt = np.random.randint(0,len(hashbrown))
        hashbrown[key_perterb[i]], hashbrown[randInt] = (0,1)
        noMove= np.append(noMove,randInt)
    return hashbrown,noMove
    
    
```

```python
# Implement FLY bloom filter

def present_odor(odor,wts):
    response = np.dot(wts,odor)/k
    wt_n = wts
    wt_n[odor==1] = wt_n[odor==1]*.5
    wt_n[odor==0] = wt_n[odor==0] + (1- wt_n[odor==0])*.2
    #wt_n[odor==0] = wt_n[odor==0]*3
    #wt_n[wt_n>1] = 1
    return response, wt_n
```

```python
n = 20
n_pert = 4
rep = 100
converge_weight = .887
# n_odor = np.shape(F)[1]

my_resp = np.zeros((n+1,rep))
for j in range(rep):
    S = outHash[np.random.choice(nIn-1,n-1,replace=False)]
    F = np.append(So,np.transpose(S),axis=1)
    _S,_ = perterb_hash(Sop,n_pert)
    _S = np.expand_dims(_S,1)
    F = np.append(F,_S,axis=1)

    w = np.ones(np.shape(F)[0])*converge_weight
    for i in range(n+1):
        my_resp[i,j],w = present_odor(F[:,i],w)
```

```python
def timing_bloom(S,O=10,q=None,bloom=None):
    # S, input set
    # O, timing max
    # q, query
    T_ = 0
    if np.any(S):
        n_item,n_dim = np.shape(S)
        bloom = np.ones(n_dim)*np.nan
        for item in range(n_item):
            expire_val = O-T_  # 10,
            bloom[bloom==expire_val] = np.nan
            bloom[S[item]==1] = T_
            T_ +=1
            if T_>O: 
                T_=0
    if np.any(q):
        this_query = bloom[q==1] # might want to loop for multiple queries
        return ~np.isnan(this_query)
    else:
        return bloom
    
def stable_bloom(S,O=10,Om = 20,q=None):
    # S, input set
    # O, timing value
    # Om, max value
    # q, query
    if np.any(S):
        n_item,n_dim = np.shape(S)
        bloom = np.ones(n_dim)*0
        for item in range(n_item):
            bloom[bloom>0] -=1 #decay factor
            bloom[S[item]==1] += O # adding factor
            bloom[bloom>Om] = Om # top out
    if np.any(q):
        this_query = bloom[q==1] # might want to loop for multiple queries
        return this_query>0
    else:
        return bloom
def exact_cylinder(S,q,time=14,space=5):
    S = S[-time:]
    n_item,n_dim = np.shape(S)
    for item in range(n_item):
        this_ham = hamming_distance(S[item],q)
        if this_ham<=space:
            if this_ham==0:
                print('Exact Match')
            return True
    else:
        return False
    
def exact_cone(S,q,space=10):
    n_item,n_dim = np.shape(S)
    for item in range(n_item):
        this_ham = hamming_distance(S[-item-1],q) # minus sign is key here
        if this_ham<=space:
            if this_ham==0:
                print('Exact Match')
            return True
        space -=1
        if space==-1:
            return False
    else:
        return False
def get_distances(S,q):
    n_item,n_dim = np.shape(S)
    d = np.zeros(n_item)
    for i in range(n_item):
        d[i] = hamming_distance(S[i],q)
    return d
hnorm = lambda x: (x-32.32)/32.32
def d_wt(d,degree=-1):
    wts = (np.arange(len(d))+1.0)**degree
    return np.dot(wts,d[-1:None:-1]) # convolution order
```
